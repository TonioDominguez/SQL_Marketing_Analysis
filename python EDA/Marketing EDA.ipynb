<h3 style="color: #6D183F;">N Clients per Education</h3>

```python
sns.set_style("whitegrid")
sns.barplot(data=income_edu, x="education", y="n_of_clients", palette="rocket")
plt.xlabel("Education Level")
plt.ylabel("Nº of clients")
plt.title("Nº of clients per Education")
plt.show()
```

<h3 style="color: #6D183F;">Average income per education</h3>

```python
sns.set_style("whitegrid")
sns.pointplot(data=income_edu, x="education", y="average_income", palette="rocket", markers="x", scale=2)
plt.xlabel("Education Level")
plt.ylabel("Average Income")
plt.title("Average Income per Education")
plt.show()
```
